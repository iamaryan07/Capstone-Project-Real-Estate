```python
import plotly.express as px
```

```python
import json
```

```python
sector_avg_price = df.groupby('Sector')['Price'].mean().reset_index()
```

```python
sector_analysis = df.groupby('Sector').agg({
    'Price': 'mean',
    'Built Up Area': 'mean'
}).reset_index()

sectors = sector_analysis['Sector'].tolist()
avg_prices = [round(price, 1) for price in sector_analysis['Price']]
avg_areas = sector_analysis['Built Up Area'].round(0).astype('int').tolist()

print("sectors:", sectors)
print("avg_price:", avg_prices)
print("avg_area:", avg_areas)
```

```python
with open(r'C:\Users\aryan\Desktop\Capstone Project\gurgaon-streamlit-app\map.geojson', 'r') as f:
        geojson_data = json.load(f)
```

```python
# --- Step 1: Create the map using the NEW function name ---
fig = px.choropleth_map(
    sector_avg_price,
    geojson=geojson_data,
    locations='Sector',
    featureidkey='properties.name',
    color='Price',
    color_continuous_scale="YlOrRd",
    map_style="carto-darkmatter",
    zoom=10,
    center={"lat": 28.4595, "lon": 77.0266},
    opacity=0.7,
    labels={'Price': 'Average Price (Cr)'}
)

# --- Step 2: Update the hover template ---
fig.update_traces(
    hovertemplate='<b>%{location}</b><br><br>Average Price: %{z:.1f} Cr<extra></extra>'
)

# --- Step 3: Final layout updates ---
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
