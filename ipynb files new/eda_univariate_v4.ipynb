```python
sector_counts = df['Sector'].value_counts()
```

```python
sector_counts
```

```python
def categorize_sector_frequency(sector):
    count = sector_counts.get(sector, 0)
    if count >= 300:
        return 'Very High'
    if (count < 300) & (count >= 100):
        return 'High'
    if (count < 100) & (count >= 20):
        return 'Medium'
    if (count < 20) & (count > 1):
        return 'Low'
    if count == 1:
        return 'Individual'
```

```python
df['Sector'].apply(categorize_sector_frequency).value_counts()
```

**Observation**

1. 52% sectors belong to Very High i.e. more than 300
2. 26% sectors belong to High i.e. less than 300  and more than 100
3. 19% sectors belong to Medium i.e. less than 100  and more than 20
4. 1.7 sectors belong to Low i.e. less than 20  and more than 1
5. 0.05 sectors belong to Individual i.e. equal to 1

```python
skewness = df['Price'].skew()
kurtosis = df['Price'].kurtosis()

print('Skewness: ', skewness)
print('Kurtosis: ', kurtosis)
```

```python
quartiles = df['Price'].quantile([0.01, 0.05, 0.5, 0.95, 0.99])

quartiles
```

```python
Q1 = df['Price'].describe()['25%']
Q3 = df['Price'].describe()['75%']
IQR = Q3 - Q1

print('Q1:  ', Q1)
print('Q3: ', Q3)
print('IQR: ', IQR)
```

```python
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

print('Lower Bound: ', lower_bound)
print('Upper Bound: ', upper_bound)
```

```python
outlier_df = df[(df['Price'] > upper_bound)]
outlier_df.shape
```

```python
outlier_df.describe()
```

*Outliers Analysis (using IQR method):*

- Based on the IQR method, there are 33 properties considered as outliers.
