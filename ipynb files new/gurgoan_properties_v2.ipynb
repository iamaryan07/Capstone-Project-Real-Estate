```python
def extract_area_values(area_text):
    super_area = built_area = carpet_area = np.nan

    if pd.isna(area_text):
        return pd.Series([super_area, built_area, carpet_area])

    # Extract all brackets like: (94.76 sq.m.) or (850 sq.ft.)
    matches = re.findall(r'([Ss]uper\s*Built\s*up\s*area|[Bb]uilt\s*[Uu]p\s*area|[Cc]arpet\s*area)[^()]*\(([\d.]+)\s*(sq\.m\.|sq\.ft\.)\)', area_text)

    # Go through matches and assign accordingly
    for label, val, unit in matches:
        val = float(val)
        if unit == 'sq.m.':
            val = round(val * 10.7639, 2)
        # Map to correct column
        if 'super' in label.lower():
            super_area = val
        elif 'built' in label.lower():
            built_area = val
        elif 'carpet' in label.lower():
            carpet_area = val

    return pd.Series([super_area, built_area, carpet_area])

# Apply function
df[['Super Built Up Area', 'Built Up Area', 'Carpet Area']] = df['Area'].apply(extract_area_values)
```

```python
def count_extra_rooms(config):
    if pd.isna(config):
        return 0
    
    config = config.lower()
    
    standard = ['bedroom', 'bathroom', 'balcony']
    
    if 'with' in config:
        parts = config.split('with', 1)
        extra = parts[1]
    else:
        extra = ','.join(config.split(',')[3:])
        
    room_list = [room.strip() for room in extra.split(',') if room.strip()]
    
    count = 0
    
    for room in room_list:
        if all(std not in room for std in standard) and room != '':
            count += 1
            
    return count

df['Extra Room'] = df['Configuration'].apply(count_extra_rooms)
```

```python
place_to_category_map = {}

for category, place_list in grouped_places.items():
    for place in place_list:
        place_to_category_map[place] = category
```

```python
def get_categories_for_row(places_list):
  """Looks up each place and returns a list of its categories."""
  if not isinstance(places_list, list):
    return [] # Return an empty list if the cell is not a list

  # For each place, find its category, or use 'Uncategorized' if not found
  return [place_to_category_map.get(place, 'Uncategorized') for place in places_list]

# Apply the function to create the new column
df['Category_Places'] = df['Places'].apply(get_categories_for_row)
```
