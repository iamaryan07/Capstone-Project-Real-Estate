```python
# Mapping of society names to sector numbers
society_to_sector = {
    'ramprastha city': 'sector 37',
    'ansals highland park': 'sector 103',
    'hero homes': 'sector 104',
    'the cocoon': 'sector 50',
    'rattan vihar': 'sector 15',
    'rajendra park': 'sector 105',
    'experion heartsong': 'sector 108',
    'smart world one': 'sector 68'
}

# Replace 'dwarka expressway' in 'sector' with the correct sector number
df['sector'] = df.apply(
    lambda row: society_to_sector.get(row['society'].strip().lower(), row['sector'])
    if row['sector'].strip().lower() == 'dwarka expressway' else row['sector'],
    axis=1
)
```

```python
df['sector'] = df['sector'].replace('gwal pahari', 'sector 2')
```

```python
df['sector'] = df['sector'].replace('dwarka expressway', 'sector 105')
```

```python
df['sector'] = df['sector'].replace('manesar', 'sector 1')
```

```python
# df['sector'] = (df['sector'].str.split(' ').str.get(1).str.strip().astype(int))
```

```python
society_sector_map = {
    'eldeco accolade': 2,
    'gls arawali homes': 4,
    'signature global park': 36,
    'hcbs sports ville': 2,
    'mvn athens': 5,
    'breez global heights': 33,
    'breez global hill view': 33,
    'trisara our homes 3': 6,
    'ashiana mulberry': 2,
    'ashiana anmol': 33,
    'silverglades the melia': 35,
    'signature the serenas': 36,
    'tulsiani easy in homes': 35,
    'central park flower valley': 32,
    'bhemerlei': 48,
    'global height': 33,
    'kamal residence': 67,
    'sneh vihar': 67,
    'g l s arawali homes': 4,
    'raheja maheshwara': 14,
    'prem kunj': 4,
    'micasa sec 68': 68
}
```

```python
def update_sector(row):
    society = row['society'].strip().lower()
    if row['sector'].strip().lower() == 'sohna road' and society in society_sector_map:
        return society_sector_map[society]
    else:
        return row['sector']

df['sector'] = df.apply(update_sector, axis=1)
```

```python
# List of possible sectors
possible_sectors = ['sector 33', 'sector 38', 'sector 47', 'sector 48', 'sector 50']

# Create a mask for the relevant rows
mask = (
    (df['sector'].str.strip().str.lower() == 'sohna road') &
    (df['society'].str.strip().str.lower() == 'independent') &
    (df['property_type'].str.strip().str.lower() == 'house')  # Adjust if needed
)

# Find the indices of the rows to update
indices = df[mask].index

# Randomly select sectors for each row
np.random.seed(42)  # For reproducibility; remove or change for different results
random_sectors = np.random.choice(possible_sectors, size=len(indices), replace=True)

# Assign the random sectors to the selected rows
df.loc[indices, 'sector'] = random_sectors
```

```python
df['sector'] = df['sector'].astype(str).str.extract(r'(\d+)', expand=False)
```

```python
# This will extract only the leading digits from the sector column (e.g., '70a' -> '70')
df['sector'] = df['sector'].astype(str).str.extract(r'(\d+)', expand=False)
```

```python
# Step 1: Convert all to string and extract digits
df['sector'] = df['sector'].astype(str).str.extract(r'(\d+)', expand=False)

# Step 2: Convert to integer
df['sector'] = df['sector'].astype(int)
```

```python
def categorize_luxury(score):
    if 0 <= score < 50:
        return 'Low'
    elif 50 <= score < 140:
        return 'Medium'
    elif 140 <= score <= 175:
        return 'High'
    else:
        return None
```

```python
train_df['luxury_category'] = train_df['luxury_score'].apply(categorize_luxury)
```

**floorNum**

```python
sns.boxplot(df['floorNum'], orient= 'h')
```

```python
def categorize_floor(floor):
    if 0 <= floor <= 2:
        return "Low Floor"
    elif 3 <= floor <= 10:
        return "Mid Floor"
    elif 11 <= floor <= 51:
        return "High Floor"
    else:
        return None
```

```python
train_df['floor_category'] = train_df['floorNum'].apply(categorize_floor)
```

```python
from sklearn.preprocessing import OrdinalEncoder

data_label_encoded = train_df.copy()

categorical_cols = train_df.select_dtypes(include=['object']).columns
# categorical_cols
for col in categorical_cols:
    encoder = OrdinalEncoder()
    data_label_encoded[col] = encoder.fit_transform(data_label_encoded[[col]])
    print(encoder.categories_)
    
X_label = data_label_encoded.drop('price', axis= 1)
y_label = data_label_encoded['price']
```

**Technique 1 - Correlation Analysis**

```python
sns.heatmap(data_label_encoded.corr())
```

```python
fi_df1 = data_label_encoded.corr()['price'].to_frame().reset_index().rename(columns={'index':'feature','price':'corr_coeff'})
fi_df1
```

**Technique 2 - Random Forest Feature Importance**

```python
from sklearn.ensemble import RandomForestRegressor

rf_label = RandomForestRegressor(random_state= 42)
rf_label.fit(X_label, y_label)

fi_df2 = pd.DataFrame({
    'feature': X_label.columns,
    'rf_importance': rf_label.feature_importances_
}).sort_values(by= 'rf_importance', ascending= False)
```

```python
fi_df2
```

**Technique 3 - Gradient Boosting Feature importances**

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_label = GradientBoostingRegressor()
gb_label.fit(X_label, y_label)

fi_df3 = pd.DataFrame({
    'feature': X_label.columns,
    'gb_importance': gb_label.feature_importances_
}).sort_values(by= 'gb_importance', ascending= False)

fi_df3
```

**Technique 4 - Permutation Importance**

```python
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split

X_train_label, X_test_label, y_train_label, y_test_label = train_test_split(X_label,  y_label, test_size= 0.2, random_state= 42)

rf_label = RandomForestRegressor(n_estimators= 100, random_state= 42)
rf_label.fit(X_train_label, y_train_label)

perm_importance = permutation_importance(rf_label, X_test_label, y_test_label, n_repeats= 30, random_state= 42)

fi_df4 = pd.DataFrame({
    'feature': X_label.columns,
    'permutation_importance': perm_importance.importances_mean
    }).sort_values(by='permutation_importance', ascending=False)

fi_df4
```

**Technique 5 - LASSO**

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_label)

lasso = Lasso(alpha= 0.01, random_state= 42)
lasso.fit(X_scaled, y_label)

fi_df5 = pd.DataFrame({
    'feature': X_label.columns,
    'lasso_coeff': lasso.coef_
}).sort_values(by= 'lasso_coeff', ascending= False)

fi_df5
```

**Technique 6 - RFE**

```python
from sklearn.feature_selection import RFE

estimator = RandomForestRegressor(random_state= 42)

selector_label = RFE(estimator= estimator, n_features_to_select= X_label.shape[1], step= 1)
selector_label = selector_label.fit(X_label, y_label)

selected_features = X_label.columns[selector_label.support_]

selected_coefficients = selector_label.estimator_.feature_importances_

fi_df6 = pd.DataFrame({
    'feature': selected_features,
    'rfe_score': selected_coefficients
}).sort_values(by='rfe_score', ascending=False)

fi_df6
```

**Technique 7 - Linear Regression Weights**

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_scaled, y_label)

fi_df7 = pd.DataFrame({
    'feature': X_label.columns,
    'reg_coeffs': lin_reg.coef_
}).sort_values(by='reg_coeffs', ascending=False)

fi_df7
```

**Technique 8 - SHAP**

```python
import shap

# Compute SHAP values using the trained Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_label, y_label)

explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(X_label)

# Summing the absolute SHAP values across all samples to get an overall measure of feature importance
shap_sum = np.abs(shap_values).mean(axis=0)

shap_values
```

```python
fi_df8 = pd.DataFrame({
    'feature': X_label.columns,
    'SHAP_score': np.abs(shap_values).mean(axis=0)
}).sort_values(by='SHAP_score', ascending=False)

fi_df8
```
